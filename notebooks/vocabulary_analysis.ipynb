# Vocabulary Analysis and Statistics

```python
import pandas as pd

import sys
import os

from pathlib import Path
from itertools import islice


sys.path.append(os.path.abspath('../src'))
```

```python
from tokenizer import Tokenizer
```

```python
tok = Tokenizer()



with open("../data/sample_reviews.txt",'r',encoding='utf-8') as f:
    all_reviews = f.readlines()
    
    
tok.build_vocab(all_reviews)


print(f"Total vocabulary size:b {len(tok.word_to_id)}")
```

```python
print(dict(islice(tok.word_to_id.items(),8)))


```

```python
sample_text = ["The cinematography was breathtakingly spectacular"]
encoded_ids = tok.encode(sample_text)
print(f"Encoded IDs: {encoded_ids}")

# 2. Transform IDs back to words
decoded_words = tok.decode(encoded_ids)
print(f"Decoded Words: {decoded_words}")
```

## Average UNK Rate

```python
total_unk = 0
total_tokens = 0


sample_reviews = all_reviews[:1000]

encoded_samples = tok.encode(sample_reviews)

for ids in encoded_samples:
    total_unk += ids.count(1)
    total_tokens += len(ids)
    
avg_unk_rate = (total_unk / total_tokens) * 100

print(f"Average UNK Rate: {avg_unk_rate:.2f}%")
```
